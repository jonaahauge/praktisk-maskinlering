```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sb

from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder, MinMaxScaler

import sklearn
import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Input, Dense




```

```python
df = pd.read_csv(r"..\oblig3\student_performance.csv", delimiter=";")
```

```python
targets = pd.Series(LabelEncoder().fit_transform(df['Target']))
features = df.drop('Target', axis=1)
```

```python
f_train, f_test, t_train, t_test = train_test_split(features, targets, test_size=0.2)
```

```python
scaler = MinMaxScaler()
scaler.fit(f_train)
f_train = scaler.transform(f_train)
f_test = scaler.transform(f_test)
```

```python
nn1 = Sequential()
nn1.add(Input(shape=(36,)))
nn1.add(Dense(1000, activation='relu'))
nn1.add(Dense(250, activation='relu'))
nn1.add(Dense(100, activation='relu'))
nn1.add(Dense(20, activation='relu'))
nn1.add(Dense(1, activation='sigmoid'))
```

```python
nn1.compile(optimizer='adam',
              loss='categorical_crossentropy',
              metrics=['accuracy'])
```

```python
nn1.fit(f_train, t_train, epochs=20, validation_split=0.2)
```
