```python
import pandas as pd
```

Først leser vi inn dataene og deler opp disse i training og testing deler.

```python
df = pd.read_csv("../Oblig1/iris_subset.csv")
train = df.iloc[:12,]
test = df.iloc[12:,]
print(f'Train: {train}')
print(f'Test: {test}')
```

Ut ifra hvordan dataene fra training delen ser ut, kan vi prøve å sortere disse på de individuelle kollonene for å se om dette skaper noen naturlige skiller.

```python
print(train.sort_values)
```

Basert på denne grov sorteringen kan det se ut som å sortere datasette basert på verdiene i kollone med index 2 vil gi oss gode skiller. 

```python
print(train.sort_index(axis=2))
```
