```python
import pandas as pd
```

Først leser vi inn dataene og deler opp disse i training og testing deler.

```python
df = pd.read_csv("../Oblig1/iris_subset.csv")
train = df.iloc[:12,]
test = df.iloc[12:,]
print(f'Train: {train}')
print(f'Test: {test}')
```

Ut ifra hvordan dataene fra training delen ser ut, kan vi prøve å sortere disse på de individuelle kollonene for å se om dette skaper noen naturlige skiller som kan gjøre det enklere for oss å gjøre beregningene.

```python
print(train.sort_values)
```

Basert på denne grove sorteringen kan det se ut som å sortere datasette basert på kollone[2] gi oss godt skille vi kan starte med å bruke.

```python
print(train.sort_values("3"))
```

Som vi ser av denne sorteringen får vi naturlige gruppering mellom Iris typene og vi har to grenseverdier  som vi kan bruke for å splite datasetet.

$$ Skille1:3,9-\frac{3,9-1,5}{2} = 2,7 $$
$$ Skille2:5,5-\frac{5,5-5,1}{2} = 5,3 $$

$$ H(S) = (-\frac{1}{3}\log_{2}(\frac{1}{3})) + (-\frac{1}{3}\log_{2}(\frac{1}{3})) + (-\frac{1}{3}\log_{2}(\frac{1}{3})) = 0,528 \times 3 = 1,584 $$
$$ H(S [2] < 5,3) = H(S [2] >= 5,3) = (-\frac{1}{3}\log_{2}(\frac{1}{3})) + (-\frac{2}{3}\log_{2}(\frac{2}{3})) = 0,528 + 0,39 = 0,918 $$
$$ IG(S[2]5,3) = 1,584 - \frac{2}{3} \times 0,918 - \frac{1}{3} \times 0,918 = 0,666 $$

$$ H(S [2] < 2,7) = H(S [2] >= 2,7) = (-\frac{1}{3}\log_{2}(\frac{1}{3})) + (-\frac{2}{3}\log_{2}(\frac{2}{3})) = 0,528 + 0,39 = 0,918 $$
$$ IG(S[2]2,7) = 1,584 - \frac{1}{3} \times 0,918 - \frac{2}{3} \times 0,918 = 0,666 $$

Som vi ser av disse utregningen så får vi samme IG ved begge av disse grenseverdiene og vi kan derfor bare velge en av disse da det ikke utgjøre noe forskjellig om hvilken av disse som brukes. Jeg velger derfor å splitte på 5,3 i dette tilfellet. Da får vi følgende tree etter denne delingen (se figur i vedlegg tre1.jpg). 

Som vi ser av treet får vi en helt ren bladnode til høyre i fordeling som kun inneholder "Iris-virginica", vi kan derfor se videre på noden til venstre for å vidre dele opp denne.
Hvis vi nå velger å sortere datasetet på kollone[3] så ser vi igjen at vi får et naturlig skille i datasetet som vi kan bruke for å videre splitte opp datasettet.

$$ Skille3:1,2-\frac{1,2-0,2}{2} = 0,7 $$

Siden vi kun har to gjennstående iriser å dele opp i, og denne spliten gir oss en ren fordeling i begge de to nye bladnoedene er det ingen behov for å gjøre videre utregninger å vi får dermed alle forekomster av "Iris-setosa" i den nye ventre bladnoden og alle forekomster av "Iris-versicolor" i den nye høre bladnoden vi vil dermed få følgende tre etter denne oppdelingen (se figur i vedlegg tre2.jpg).
