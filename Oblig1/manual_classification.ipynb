Først leser vi inn dataene og deler opp disse i training og testing deler.

```python
df = pd.read_csv("../Oblig1/iris_subset.csv")
train = df.iloc[:12,]
test = df.iloc[12:,]
print(f'Train: {train}')
print(f'Test: {test}')
```

Ut ifra hvordan dataene fra training delen ser ut, kan vi prøve å sortere disse på de individuelle kollonene for å se om dette skaper noen naturlige skiller som kan gjøre det mulig for oss å gjennomføre
færre beregninger.

Ref for iloc[]: https://pandas.pydata.org/docs/reference/api/pandas.DataFrame.iloc.html#pandas.DataFrame.iloc

```python
print(train.sort_values)
```

Basert på denne grov sorteringen kan det se ut som å sortere datasette basert på verdiene i kollone med index 2 vil gi oss gode skiller. 

Referanse for sort_values(): https://pandas.pydata.org/docs/reference/api/pandas.DataFrame.sort_values.html

```python
print(train.sort_values("2"))
```

Som vi kan se av sortering basert på verdiene i kolloene med index 2 får vi klare skiller i train datasetet og vi kan derfor prøve å bruke disse naturlige skillene for å gjennomføre
den manuelle klasifikasjonen av datasettet.

For å sette opp disse skillene så kan vi bruke ytterpunktene i verdiene for kollonen med index 2 mellom de naturlige skillene.

Skille_1: 3,9 - ((3,9 - 1,5) / 2) = 2,7 


Skille_2: 5,5 - ((5,5 - 5,1) / 2) = 5,3

H(S) = (-1/3 log_2(1/3))+(-1/3 log_2(1/3))+(-1/3 log_2(1/3)) = 0,528 * 3 = 1,584

Ut i fra sorteringen vi har gjort på dataene i train så ser vi at vi får brøkmessig like splitter ut i fra de skillene vi har satt opp. Alle disse inneholder 1/3 av den totale mengden data i train og vil derfor gi samme oss Information Gain som vist at utregningen under. Det er derfor  

H(S, col_2 < 2,7) = H(S, col_2 >= 2,7 and col_2 < 5,3) = H(S, col_2 >= 5,3) = 
(-1/3 log_2(1/3)) + (-2/3 log_2(2/3)) = 0,528 + 0,39 = 0,918
